```python
import pandas as pd
```

```python
data=pd.read_csv('data.csv')
data
```

```python
data.drop_duplicates(inplace=True,subset=['name'])
```

```python
name=data['name']
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
col_features = ['danceability', 'energy', 'valence', 'loudness']
X = MinMaxScaler().fit_transform(data[col_features])
kmeans = KMeans(init="k-means++",
                n_clusters=7,
                random_state=15).fit(X)
data['kmeans'] = kmeans.labels_
```

```python
data['song_name']=name
```

```python
data
```

```python
cluster=data.groupby(by=data['kmeans'])
```

```python
y=data.pop('kmeans')
x=data.drop(columns=['name','artists','id','release_date','song_name'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
# !pip install lightgbm
from lightgbm import LGBMClassifier
model=LGBMClassifier().fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
df=cluster.apply(lambda x: x.sort_values(["popularity"],ascending=False))
df.reset_index(level=0, inplace=True)
```

```python
# from keras.preprocessing.image import img_to_array
import keras.utils as image
import imutils
from keras.models import load_model
import numpy as np
import cv2
```

```python
detection_model_path = 'haarcascade_frontalface_default.xml'
emotion_model_path1 = 'model1.h5'
face_detection = cv2.CascadeClassifier(detection_model_path)
emotion_classifier = load_model(emotion_model_path1, compile=False)
EMOTIONS = ["happy","sad","neutral","angry"]
```

```python
#-------------------------------------------------------#
#                    IMPORT PACKAGES                    #
#-------------------------------------------------------#
from time import sleep
from keras.preprocessing import image
import cv2
import keras.utils as image
import numpy as np
#-------------------------------------------------------#


#-------------------------------------------------------#
#              DEFINING VARIABLES AND PATHS             #  Q
#-------------------------------------------------------#
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# class_labels = ['Angry','Happy','Neutral','Sad','Surprise']
#-------------------------------------------------------#



cap = cv2.VideoCapture(0)
while True:
    # Grab a single frame of video
    ret, frame = cap.read()
    labels = []
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray,1.3,5)

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(int(x),int(y)),(int(x+w),int(y+h)),(205,0,0),2)
        roi_gray = gray[y:y+h,x:x+w]
        roi_gray = cv2.resize(roi_gray,(48,48),interpolation=cv2.INTER_AREA)


        if np.sum([roi_gray])!=0:
            roi = roi_gray.astype('float')/255.0
            roi = image.img_to_array(roi)
            roi = np.expand_dims(roi,axis=0)

        # make a prediction on the ROI, then lookup the class

            preds = emotion_classifier.predict(roi)[0]
            print("\nprediction = ",preds)
            label=EMOTIONS[preds.argmax()]
            print("\nprediction max = ",preds.argmax())
            print("\nlabel = ",label)
            label_position = (x,y)
            cv2.putText(frame,label,label_position,cv2.FONT_HERSHEY_SIMPLEX,2,(0,255,0),3)
        else:
            cv2.putText(frame,'No Face Found',(20,60),cv2.FONT_HERSHEY_SIMPLEX,2,(0,255,0),3)
        print("\n\n")
    cv2.imshow('Emotion Detector',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
emotion_word=label
if emotion_word=='sad':
    emotion_code=0
elif emotion_word=='happpy':
    emotion_code=1
elif emotion_word=='neutral':
    emotion_code=2
elif emotion_word=='angry':
    emotion_code=3
elif emotion_word=='fear':
    emotion_code=4
elif emotion_word=='surprise':
    emotion_code=5
```

```python
def get_results(emotion_code):
  NUM_RECOMMEND=10
  happy_set=[]
  sad_set=[]
  angry_set=[]
  neutral_set=[]
  fear_set=[]
  surprise_set=[]
  if emotion_code==0:
      sad_set.append(df[df['kmeans']==0]['song_name'].head(NUM_RECOMMEND))
      return pd.DataFrame(sad_set).T
  elif emotion_code==1:
      happy_set.append(df[df['kmeans']==1]['song_name'].head(NUM_RECOMMEND))
      return pd.DataFrame(happy_set).T
  elif emotion_code==2:
      neutral_set.append(df[df['kmeans']==2]['song_name'].head(NUM_RECOMMEND))
      return pd.DataFrame(neutral_set).T
  elif emotion_code==3:
      angry_set.append(df[df['kmeans']==3]['song_name'].head(NUM_RECOMMEND))
      return pd.DataFrame(angry_set).T
  elif emotion_code==4:
      fear_set.append(df[df['kmeans']==4]['song_name'].head(NUM_RECOMMEND))
      return pd.DataFrame(fear_set).T
  elif emotion_code==5:
      surprise_set.append(df[df['kmeans']==5]['song_name'].head(NUM_RECOMMEND))
      return pd.DataFrame(surprise_set).T
```

```python
print(get_results(emotion_code))
if emotion_word=='sad':
    print('emotion detected is SAD')
elif emotion_word=='happy':
    print('emotion detected is HAPPY')
elif emotion_word=='angry':
    print('emotion detected is ANGRY')
elif emotion_word=='neutral':
    print('emotion detected is NEUTRAL')
```
